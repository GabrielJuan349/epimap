```python
# Cargar el archivo de códigos de enfermedades
epi_codes = pd.read_csv('../data/epi_codes.csv')

# Crear un diccionario para mapear cada código detallado a su código numérico
code_mapping = {}

# Procesar los códigos
for _, row in epi_codes.iterrows():
    code_num = row['Code']
    
    # Si la columna Detailed List Numbers está vacía, saltar
    if pd.isna(row['Detailed List Numbers']):
        continue
        
    # Procesar códigos individuales o en rangos
    details = str(row['Detailed List Numbers']).strip()
    
    # Dividir por comas para manejar múltiples entradas
    entries = [entry.strip() for entry in details.split(',')]
    
    for entry in entries:
        # Verificar si es un rango (contiene un guión)
        if '-' in entry:
            start, end = entry.split('-')
            
            # Para códigos simples como A00-A99
            if start[0] == end[0]:  # Misma letra inicial
                letter = start[0]
                # Verificar si hay puntos decimales
                if '.' not in start and '.' not in end:
                    try:
                        start_num = int(start[1:])
                        end_num = int(end[1:])
                        for i in range(start_num, end_num + 1):
                            code = f"{letter}{i:02d}"  # formato con ceros a la izquierda
                            code_mapping[code] = code_num
                    except ValueError:
                        # Si no se puede convertir a int, usar el rango completo
                        code_mapping[entry] = code_num
                else:
                    # Si hay decimales, usar el rango completo
                    code_mapping[entry] = code_num
        else:
            # Códigos individuales
            code_mapping[entry] = code_num

# Manejar manualmente los códigos específicos con decimales
# Agregar mapeos específicos para U07.0, U07.1 y U07.2 si aparecen en tus datos
# Por ejemplo (ajusta según corresponda):
code_mapping['U07.0'] = 1902  # Asume un código apropiado
code_mapping['U07.1'] = 1903  # Código para COVID-19
code_mapping['U07.2'] = 1903  # Otro código relacionado con COVID

# Definir una función de mapeo más robusta
def map_cause_code(cause):
    # Si ya es un código numérico, mantenerlo
    if isinstance(cause, (int, float)) or (isinstance(cause, str) and cause.isdigit()):
        return cause
    
    # Intentar mapeo directo
    if cause in code_mapping:
        return code_mapping[cause]
    
    # Si no hay mapeo directo, intentar con el código base (sin decimal)
    if '.' in cause:
        base_code = cause.split('.')[0]
        if base_code in code_mapping:
            return code_mapping[base_code]
    
    # Devolver el original si no hay mapeo
    return cause

# Aplicar el mapeo al DataFrame
df['Cause_Code'] = df['Cause'].apply(map_cause_code)

# Ahora agrupar por país, año, sexo y código numérico de causa
df = df.groupby(['Country', 'Year', 'Sex', 'Cause_Code'], as_index=False).sum().drop('Cause', axis=1)

# Ver algunas filas para verificar
df.head(5)
```
