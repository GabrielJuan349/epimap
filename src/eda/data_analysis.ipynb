```python
df = pd.read_csv('../../data/Morticd10_part6')
df
```

```python
df.drop(['SubDiv', 'List', 'Frmat', 'IM_Frmat', 'Deaths1', 'IM_Deaths1', 'IM_Deaths2', 'IM_Deaths3', 'IM_Deaths4'], axis=1, inplace=True)
df
```

```python
death_cols = ['Deaths2', 'Deaths3', 'Deaths4', 'Deaths5', 'Deaths6']

df['Deaths2-6'] = df[death_cols].fillna(0).sum(axis=1)
df.drop(death_cols, axis=1, inplace=True)
df.columns
```

```python
death_cols = [col for col in df.columns if col.startswith('Deaths') and col[-1].isdigit()]
death_cols.pop(-1)
for i in range(0, len(death_cols), 2):
    if i+1 < len(death_cols):
        col1 = death_cols[i]
        col2 = death_cols[i+1]
        new_col_name = f'Deaths{col1[6:]}-{col2[6:]}'
        df[new_col_name] = df[col1].fillna(0) + df[col2].fillna(0)

df.drop(death_cols, axis=1, inplace=True)
df.columns
```

```python
df = df.groupby(['Country', 'Year', 'Sex', 'Cause']).sum(numeric_only=True).reset_index().drop(['Admin1'], axis=1)
```

```python
# Cargar el archivo de códigos de enfermedades
epi_codes = pd.read_csv('../../data/epi_codes.csv')

# Crear un diccionario para mapear cada código detallado a su código numérico
code_mapping = {}

# Procesar los códigos
for _, row in epi_codes.iterrows():
    code_num = row['Code']
    
    # Si la columna Detailed List Numbers está vacía, saltar
    if pd.isna(row['Detailed List Numbers']):
        continue
        
    # Procesar códigos individuales o en rangos
    details = str(row['Detailed List Numbers']).strip()
    
    # Dividir por comas para manejar múltiples entradas
    entries = [entry.strip() for entry in details.split(',')]
    
    for entry in entries:
        # Verificar si es un rango (contiene un guión)
        if '-' in entry:
            start, end = entry.split('-')
            
            # Para códigos simples como A00-A99
            if start[0] == end[0]:  # Misma letra inicial
                letter = start[0]
                # Verificar si hay puntos decimales
                if '.' not in start and '.' not in end:
                    try:
                        start_num = int(start[1:])
                        end_num = int(end[1:])
                        for i in range(start_num, end_num + 1):
                            code = f"{letter}{i:02d}"  # formato con ceros a la izquierda
                            code_mapping[code] = code_num
                    except ValueError:
                        # Si no se puede convertir a int, usar el rango completo
                        code_mapping[entry] = code_num
                else:
                    # Si hay decimales, usar el rango completo
                    code_mapping[entry] = code_num
        else:
            # Códigos individuales
            code_mapping[entry] = code_num

# Manejar manualmente los códigos específicos con decimales
# Agregar mapeos específicos para U07.0, U07.1 y U07.2 si aparecen en tus datos
# Por ejemplo (ajusta según corresponda):
code_mapping['U07.0'] = 1902  # Asume un código apropiado
code_mapping['U07.1'] = 1903  # Código para COVID-19
code_mapping['U07.2'] = 1903  # Otro código relacionado con COVID

# Definir una función de mapeo más robusta
def map_cause_code(cause):
    # Si ya es un código numérico, mantenerlo
    if isinstance(cause, (int, float)) or (isinstance(cause, str) and cause.isdigit()):
        return cause
    
    # Intentar mapeo directo
    if cause[:3] in code_mapping:
        return code_mapping[cause[:3]]
    
    # Si no hay mapeo directo, intentar con el código base (sin decimal)
    if '.' in cause:
        base_code = cause.split('.')[0]
        if base_code in code_mapping:
            return code_mapping[base_code]
    
    # Devolver el original si no hay mapeo
    return cause

# Aplicar el mapeo al DataFrame
df['Cause_Code'] = df['Cause'].apply(map_cause_code)

# Identificar las filas donde el mapeo no funcionó
# Primero convertimos Cause_Code a string para poder comparar
df['Cause_Code_Str'] = df['Cause_Code'].astype(str)
df['Cause_Str'] = df['Cause'].astype(str)

# Encontrar registros sin mapeo (donde Cause == Cause_Code)
unmapped = df['Cause_Str'] == df['Cause_Code_Str']
print(f"Filas con códigos sin mapear: {unmapped.sum()} de {len(df)}")

# Ver algunos ejemplos de códigos sin mapear
if unmapped.sum() > 0:
    print("Ejemplos de códigos sin mapear:")
    print(df[unmapped]['Cause'].value_counts().head(10))

# Eliminar las filas sin mapeo
df_mapped = df[~unmapped]
print(f"Filas después de eliminar códigos sin mapear: {len(df_mapped)} de {len(df)}")

# Eliminar columnas auxiliares
df_mapped = df_mapped.drop(['Cause_Str', 'Cause_Code_Str'], axis=1)

# Ahora agrupar por país, año, sexo y código numérico de causa
df = df_mapped.groupby(['Country', 'Year', 'Sex', 'Cause_Code'], as_index=False).sum().drop('Cause', axis=1)
df['Cause_Code'] = df['Cause_Code'].astype(int)

# Ver algunas filas para verificar
df.head(5)
```

```python
pop_cols = [col for col in df_pop.columns if col.startswith('Pop') and col[-1].isdigit()]
pop_cols.pop(-1)
for i in range(0, len(pop_cols), 2):
    if i+1 < len(pop_cols):
        col1 = pop_cols[i]
        col2 = pop_cols[i+1]
        new_col_name = f'Pop{col1[3:]}-{col2[3:]}'
        df_pop[new_col_name] = df_pop[col1].fillna(0) + df_pop[col2].fillna(0)

df_pop.drop(pop_cols, axis=1, inplace=True)
df_pop.columns
```

# Dataframe Fusion

```python
# Fusionar df y df_pop basándose en Country, Year y Sex
df_merged = pd.merge(
    left=df,
    right=df_pop,
    on=['Country', 'Year', 'Sex'],
    how='left'
)

# Verificar el resultado
print(f"Número de filas en df original: {len(df)}")
print(f"Número de filas en df_merged: {len(df_merged)}")

# Ver las columnas de población añadidas
pop_columns = [col for col in df_merged.columns if 'Pop' in col]
print(f"Columnas de población añadidas: {pop_columns}")

# Verificar si hay valores NaN en las columnas de población
nan_count = df_merged[pop_columns].isna().sum().sum()
if nan_count > 0:
    print(f"¡Atención! {nan_count} valores NaN en columnas de población")
    
    # Ver qué países/años faltan en df_pop
    missing_data = df_merged[df_merged[pop_columns[0]].isna()]
    print(f"Ejemplos de registros sin datos de población:")
    print(missing_data[['Country', 'Year', 'Sex']].head(5))

# Actualizar el dataframe principal
df = df_merged

# Mostrar las primeras filas para verificar la fusión
df.head()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
# Identificar todas las columnas de muertes
deaths_columns = [col for col in df.columns if 'Deaths' in col]

# Para cada columna de muertes, calcular la proporción
for deaths_col in deaths_columns:
    # Obtener el sufijo numérico (ej: "0-4" de "Deaths0-4")
    suffix = deaths_col.replace('Deaths', '')
    
    # Encontrar la columna de población correspondiente
    pop_col = f'Pop{suffix}'
    
    # Verificar que la columna de población existe
    if pop_col in df.columns:
        # Calcular la proporción (muertes / población)
        # Evitar divisiones por cero
        df[deaths_col] = df[deaths_col] / df[pop_col].replace(0, np.nan)
        df[deaths_col] = df[deaths_col].replace(np.nan, 0)
        
        # Opcional: renombrar la columna para indicar que ahora es una tasa
        # df = df.rename(columns={deaths_col: f'Rate{suffix}'})
    else:
        print(f"No se encontró columna de población para {deaths_col}")

# Mostrar las primeras filas para verificar el cambio
df.head()

# Ver estadísticas descriptivas de las nuevas proporciones
df[deaths_columns].describe()
```
